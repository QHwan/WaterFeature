```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
f = np.load('../data.npz')
x = f['feature']
y = f['propensity']
```

```python
print(x.shape, y.shape)
```

```python
x = x.reshape(-1, 12)
y = y.flatten()
```

```python
sns.distplot(y)
```

```python
fig, ax = plt.subplots(nrows=3, ncols=4, figsize=(18, 12))
idx = 0
for i in range(3):
    for j in range(4):
        sns.distplot(x[:,idx], ax=ax[i,j])
        idx += 1
```

```python
scaler = StandardScaler()
data = np.concatenate((x, y.reshape(-1,1)), axis=1)
scaler.fit(data)
data_norm = scaler.transform(data)
x_norm = data_norm[:,:-1]
y_norm = data_norm[:,-1]
```

```python
from sklearn.svm import NuSVR
from sklearn.tree import DecisionTreeRegressor

svr = NuSVR()
svr.fit(x_norm, y_norm)
y_pred = svr.predict(x_norm)
```

```python
x = np.linspace(np.min(y_norm), np.max(y_norm), 1000)
plt.plot(x, x)
plt.scatter(y_norm, y_pred, s=5, alpha=.5)
```
